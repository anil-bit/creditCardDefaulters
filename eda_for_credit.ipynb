```python
x = data.drop(["default payment next month"],axis = 1)
```

```python
x.columns
```

```python
y = data["default payment next month"]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.33,random_state=42)
```

```python
x_train
```

```python
from sklearn.preprocessing import StandardScaler
train_scaler = StandardScaler()
test_scaler = StandardScaler()
```

```python
scaled_train = train_scaler.fit_transform(x_train)
scaled_test = test_scaler.fit_transform(x_test)
```

```python
scaled_train_df = pd.DataFrame(scaled_train,columns=x_train.columns,index = x_train.index)
```

```python
scaled_train_df
```

```python
scaled_test_df = pd.DataFrame(scaled_test,columns=x_test.columns,index = x_test.index)
```

```python
scaled_test_df
```

#first we use navies bia therome

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gnb = GaussianNB()
```

```python
pred_y = gnb.fit(scaled_train_df,y_train).predict(scaled_test_df)
```

```python
from sklearn.metrics import accuracy_score
```

```python
ac = accuracy_score(y_test,pred_y)
```

```python
ac
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {"var_smoothing": [1e-9,0.1, 0.001, 0.5,0.05,0.01,1e-8,1e-7,1e-6,1e-10,1e-11]}
```

```python
grid = GridSearchCV(estimator=gnb,param_grid=param_grid,cv=5,verbose=3)
```

```python
grid.fit(scaled_train_df,y_train)
```

```python
print(grid.best_estimator_)
```

```python
pred_y_new = grid.best_estimator_.fit(scaled_train_df,y_train).predict(scaled_test_df)
```

```python
ac_new = accuracy_score(y_test,pred_y_new)
```

```python
ac_new
```

```python
ac
```
